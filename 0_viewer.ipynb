```python
def plot(p): 
    img = nib.load(p).get_fdata()
    x = int(img.shape[0]/2)
    y = int(img.shape[1]/2)
    z = int(img.shape[2]/2)
    imx = img[x,:,:]
    imy = img[:,y,]
    imz = img[:,:,z]
    fig = plt.figure()
    ax1 = fig.add_subplot(1, 3, 1)
    ax1.axis('off')
    ax1.imshow(imx)
    ax2 = fig.add_subplot(1, 3, 2)
    ax2.axis('off')
    ax2.imshow(imy)
    ax3 = fig.add_subplot(1, 3, 3)
    ax3.axis('off')
    ax3.imshow(imz)
    plt.show()
```

```python
p = './data/ADNIData/ADNI_002_S_0413_MR_MPR____N3__Scaled_2_Br_20081001114937668_S14782_I118675.nii'
plot(p)
```
