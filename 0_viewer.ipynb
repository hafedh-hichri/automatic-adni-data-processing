```python
def plot(p): 
    img = nib.load(p).get_fdata()
    x = int(img.shape[0]/2)
    y = int(img.shape[1]/2)
    z = int(img.shape[2]/2)
    imx = img[x,:,:]
    imy = img[:,y,]
    imz = img[:,:,z]
    plt.imshow(imx)
    plt.show()
    plt.imshow(imy)
    plt.show()
    plt.imshow(imz)
    plt.show()
```
